```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
data = pd.read_csv(url)
```

```python
print('dataset preview')
print(data.head())
```

```python
print('basic information')
print(data.info())
```

```python
print('summary statistics')
print(data.describe())
```

```python
print('unique species')
print(data['species'].unique())
```

```python
# Data visualization: Histogram of Sepal Length
plt.figure(figsize=(8, 5))
plt.hist(data['sepal_length'], bins=15, color='skyblue', edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatter plot for sepal length vs sepal widht 
plt.figure(figsize=(8,5))
for species in data['species'].unique():
    subset = data[data['species']==species]
    plt.scatter(subset['sepal_length'],subset['sepal_width'],label=species)
plt.title('sepal length vs sepal width by species')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.legend()
plt.show()

    
```

```python
#correlation heatmap
import seaborn as sns

plt.figure(figsize=(8,5))
correlation_matrix  = data.iloc[:, :-1].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
